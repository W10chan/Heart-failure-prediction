```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# データの読み込み
data = pd.read_csv('drive/MyDrive/心不全予測/train.csv')

# 特徴量とターゲット変数の分割
X = data.drop(columns=['id','target'], axis=1)
y = data['target']

# 訓練データとテストデータに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)

# 弱学習器としてDecisionTreeClassifierを使用
base_estimator = DecisionTreeClassifier(max_depth=1)

# AdaBoost分類器を定義
adaboost_clf = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=50, random_state=42)

# モデルを訓練
adaboost_clf.fit(X_train, y_train)

# テストデータで予測
y_pred = adaboost_clf.predict(X_test)

# 精度を評価
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# テストデータのCSVファイルを読み込む
test_df = pd.read_csv('drive/MyDrive/心不全予測/test.csv')

# テストデータから特徴量を取得
X_test = test_df.drop('id', axis=1)

# テストデータでの予測
y_pred = adaboost_clf.predict(X_test)
y_pred = [1 if p >= 0.5 else 0 for p in y_pred]

# 予測結果を持つDataFrameを作成
submission_df = pd.DataFrame({
    'index': test_df['id'],  # テストデータのインデックス
    'prediction': y_pred        # 予測結果
})

# CSVファイルとして保存（ヘッダ無し）
submission_df.to_csv('drive/MyDrive/心不全予測/AdaBoost分類器_1.csv', header=False, index=False)
```
